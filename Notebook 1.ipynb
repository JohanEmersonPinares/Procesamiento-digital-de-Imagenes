Crear UNA funci√≥n para transformar una imagen en RGB a grises y luego al formato binario, lee los siguientes pasos:

Si ingresas una imagen en RGB a la funci√≥n debe regresar la conversi√≥n a grises y luego el binario.
Si ingresas una imagen en grises a la funci√≥n debe retornar tan solo la imagen binaria.
Crea dos ejemplos probando cada caso, una imagen RGB y luego una en grises.

```
import cv2
```

Imagen RGB a Gris y luego a Binario

```
imagen_pg1=cv2.imread("/work/pc1.jpeg")
imagen_rgb = cv2.cvtColor(imagen_pg1, cv2.COLOR_BGR2RGB)
plt.imshow(imagen_rgb)

```

```
def convertirrgbgris(imagen,umbral):
    d=len(imagen.shape)
    if d==3:
        R=imagen[:,:,0]
        G=imagen[:,:,1]
        B=imagen[:,:,2]
        imagen_gris=0.299*R+0.587*G+0.114*B
        imagen_binaria=imagen_gris > umbral 
        return (imagen_gris.astype(np.uint8),imagen_binaria.astype(np.ubyte))      
    else:
        imagen_binaria = imagen > umbral 
        return imagen_binaria.astype(np.ubyte)
```

```
(imggris,imggbin)= convertirrgbgris(imagen_rgb,126)

plt.gray()

f, (ax1, ax2) = plt.subplots(1, 2)
plt.xticks([])
plt.yticks([])
ax1.imshow(imggris)
ax2.imshow(imggbin)
ax1.set_title("Gris")
ax2.set_title("Binaria")
plt.show()
```
